```python
import requests #para accedere a la pagina web 
from bs4 import BeautifulSoup
url = "https://www.carrefour.es/supermercado/el-mercado/quesos/todos-los-quesos/N-ndzi6a/c?No=0&Nr%3DAND(product.shopCodes%3A004320%2Cproduct.salepointWithActivePrice_004320%3A1%2COR(product.siteId%3AbasicSite))OR)"
response = requests.get(url,  headers={'User-Agent': 'Mozilla/5.0'})
print(bool(response))
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
todosquesos = soup.find_all('li', class_ = 'product-card-list__item') # all types of cheese from the page 
todosquesos
```

```python
import numpy as np 
listaprecios = []
for i in range(len(todosquesos)):
    try: 
    
        try: #looks for a price with a discount
            x = todosquesos[i].find('span', class_ = "product-card__price--current").text.strip()
            listaprecios.append(x)
        except: #if there is no found puts nan
            x = np.nan
            listaprecios.append(np.nan)
    except:
        listaprecios.append(np.nan) #there are some blocks on the page that contains ads and other elements that produce an error
            
            
listaprecios 
```

```python
#first page of the website
listafinal = []

for i in range(len(todosquesos)):
    try:
        name = todosquesos[i].find('h2', class_ = 'product-card__title').text.split('\n')[1].split('          ')[1]
        pricenormal = todosquesos[i].find('div', class_= 'product-card__prices').text.split('\n')[-2].strip()
        
        
        
        try: #busca precio con descuento
            x = todosquesos[i].find('span', class_ = "product-card__price--strikethrough").text.strip()
            listafinal.append([name, pricenormal,x])
        except: #sino hay pone precio normal 
            x = np.nan
            listafinal.append([name, pricenormal,x])
            
    except:
        pass
        
listafinal 
```

```python
import pandas as pd
df = pd.DataFrame(listafinal, columns = ["Name",'Current Price', 'Price before discount'])
df
```
