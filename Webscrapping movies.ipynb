```python
import requests 
from bs4 import BeautifulSoup
```

```python
url = "https://www.filmaffinity.com/es/topcat.php?id=new_th_es" #webpage
response = requests.get(url)
print(bool(response))
soup = BeautifulSoup(response.text, "html.parser")
```

```python
lista2 = []
for i in range(40):
    posicion = soup.find('div', class_='cat-list').find_all('div', class_ = 'top-movie')[i].find("div", class_ = "pos").text
    titulo = soup.find('div', class_='cat-list').find_all('div', class_ = 'top-movie')[i].find("div", class_ = "movie-card movie-card-22").find("a")['title']
    duracion = soup.find('div', class_='cat-list').find_all('div', class_ = 'top-movie')[i].find("div", class_ = "duration").text[2:5]
    ano = soup.find('div', class_='cat-list').find_all('div', class_ = 'top-movie')[i].find("div", class_ = "mc-data").text[1:5]
    director = soup.find('div', class_='cat-list').find_all('div', class_ = 'top-movie')[i].find("div", class_ = "credits").text
    genero = soup.find('div', class_='cat-list').find_all('div', class_ = 'top-movie')[i].find('a', class_ = 'genre').text
    rating = soup.find('div', class_='cat-list').find_all('div', class_ = 'top-movie')[i].find('div', class_ = 'avg-rating').text
    numero = soup.find('div', class_='cat-list').find_all('div', class_ = 'top-movie')[i].find('div', class_ = 'rat-count').text[1:7]
    listaactores = []
    for j in range(len(soup.find('div', class_='cat-list').find_all('div', class_ = 'top-movie')[i].find("div", class_ = "cast").find_all('span', class_ = 'nb'))):
        actor = soup.find('div', class_='cat-list').find_all('div', class_ = 'top-movie')[i].find("div", class_ = "cast").find_all('span', class_ = 'nb')[j].text
        listaactores.append(actor)

    lista2.append([posicion,titulo,duracion,ano,director,genero,rating,numero,listaactores])
```

```python
import pandas as pd
df = pd.DataFrame(lista2, columns = ["Position","Title","Duration",'Year', 'Director','Genre','Rating','Number of reviews','Actors'])
df
```
